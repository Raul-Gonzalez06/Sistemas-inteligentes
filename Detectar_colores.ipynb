```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 
```

```python
img=cv2.imread('IMG/Suavizado.png')
img=cv2.resize(img,(400,400))
img=cv2.GaussianBlur(img,(9,9),0)
#img=cv2.medianBlur(img,7)
cv2.imshow('Figura',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#H100-125 S V
Azul_B = np.array([80,100,100],np.uint8)
Azul_A = np.array([130,255,255],np.uint8)
hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
mask =cv2.inRange(hsv,Azul_B,Azul_A)
contorno,_ =cv2.findContours(mask,cv2.RETR_TREE,cv2.CHAIN_APPROX_NONE)
for i in range(len(contorno)):
 cv2.drawContours(img,contorno,i,(255,0,0),3)
cv2.imwrite('IMG/Contorno_Azul.png',img)

#cv2.imshow('mask',hsv)
cv2.imshow('mask',mask)
cv2.imshow('frame',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
