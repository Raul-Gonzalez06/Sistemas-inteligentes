```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 
```

```python
img1=cv2.imread('IMG/4.jpg')
img=cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY)
img=cv2.resize(img,(500,500))
cv2.imshow('Figura',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Contorno de Imagenes 
* Escala de blanco y negor 
* Delimitar el contorno 
<table border="1">
    <thead>
        <tr>
            <th>Modo de Contorno</th>
            <th>Descripción</th>
        </tr>
    </thead>
    <tbody>
        <tr>
            <td><code>cv2.RETR_LIST</code></td>
            <td>Extrae todos los contornos sin jerarquía.</td>
        </tr>
        <tr>
            <td><code>cv2.RETR_EXTERNAL</code></td>
            <td>Solo extrae los contornos externos.</td>
        </tr>
        <tr>
            <td><code>cv2.RETR_TREE</code></td>
            <td>Extrae todos los contornos y mantiene la jerarquía completa.</td>
        </tr>
        <tr>
            <td><code>cv2.RETR_CCOMP</code></td>
            <td>Extrae todos los contornos, pero organiza la jerarquía en dos niveles (externos e internos).</td>
        </tr>
    </tbody>
</table>

```python

_,unbra=cv2.threshold(img,150,255,cv2.THRESH_BINARY_INV,cv2.THRESH_OTSU)
RETR_LIST,_ =cv2.findContours(unbra,cv2.RETR_LIST,cv2.CHAIN_APPROX_NONE)
RETR_EXTERNAL,_ =cv2.findContours(unbra,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)
RETR_TREE,_ =cv2.findContours(unbra,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
RETR_CCOMP,_ =cv2.findContours(unbra,cv2.RETR_CCOMP,cv2.CHAIN_APPROX_SIMPLE)

print(len(RETR_LIST))
print(len(RETR_EXTERNAL))
print(len(RETR_TREE))
print(len(RETR_CCOMP))
```

* Dibujar el contorno 


```python

f=[RETR_LIST,RETR_EXTERNAL,RETR_TREE,RETR_CCOMP]
t=['RETR_LIST','RETR_EXTERNAL','RETR_TREE','RETR_CCOMP']
for i in range(len(f)):
   cv2.drawContours(unbra,f[i],-1,(0,255,255),3)
   img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
   plt.subplot(2,2,i+1)
   plt.imshow(img,vmin=0,vmax=255)
   plt.title(t[i])
   plt.xticks([])
   plt.yticks([])
plt.show()
```

```python
_,unbra=cv2.threshold(img1,150,255,cv2.THRESH_BINARY,cv2.THRESH_OTSU)
RETR_LIST,_ =cv2.findContours(unbra,cv2.RETR_LIST,cv2.CHAIN_APPROX_NONE)
RETR_EXTERNAL,_ =cv2.findContours(unbra,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)
RETR_TREE,_ =cv2.findContours(unbra,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
RETR_CCOMP,_ =cv2.findContours(unbra,cv2.RETR_CCOMP,cv2.CHAIN_APPROX_SIMPLE)
print(len(RETR_EXTERNAL))

```

```python
f=[RETR_LIST,RETR_EXTERNAL,RETR_TREE,RETR_CCOMP]
t=['RETR_LIST','RETR_EXTERNAL','RETR_TREE','RETR_CCOMP']

for i in range(len(f)):
   cv2.drawContours(img1,f[i],-1,(255,255,0),3)
   img1=cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
   plt.subplot(2,2,i+1)
   plt.imshow(img1,vmin=0,vmax=255)
   plt.title(t[i])
   plt.xticks([])
   plt.yticks([])
plt.show()
```
