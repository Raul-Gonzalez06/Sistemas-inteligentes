```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 
```

```python
img=cv2.imread("IMG/Suavizado.png")
img_size=400
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
img=cv2.resize(img,(img_size,img_size))
cv2.imshow('figura',img)
cv2.waitKey(0)
cv2.destroyAllWindows()

```

# Suavizado de imagenes
* Aplicando filtros (mascaras) 

```python
kernel=np.ones((5,5),np.float32)/25
kernel1=np.ones((3,3),np.float32)/3
kernel2=np.ones((8,8),np.float32)/64
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
Conv=cv2.filter2D(img,-1,kernel2)
Prom=cv2.blur(img,(6,6))
Gaus=cv2.GaussianBlur(img,(3,3),0)
Medi=cv2.medianBlur(img,9)
con_medi=cv2.filter2D(Medi,-1,kernel1)
Gaus_Gaus=cv2.GaussianBlur(Gaus,(3,3),0)
f=[Conv,Prom,Gaus,Medi,con_medi,Gaus_Gaus]
t=['Convolucion','Promedio','Gaussiano','Mediana','Convolucion_Mediana','2_Gaussiano']

for i in range(len(f)):
   plt.subplot(2,3,i+1)
   plt.imshow(f[i],vmin=0,vmax=255)
   plt.title(t[i])
   plt.xticks([])
   plt.yticks([])
plt.show()
```
